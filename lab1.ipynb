```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generar datos de ejemplo
np.random.seed(0)
# Supongamos un incremento lineal en el precio de la acción con el tiempo, con ruido aleatorio
# Tomemos un rango de valores de 2010 a 2023 (168 meses en total)
months = np.arange(1, 169)
precio_accion = 1.5 * months + np.random.normal(0, 20, len(months))

# Ajustar el precio de la acción para asegurarse de que no sea menor que 2
precio_accion = np.maximum(5, precio_accion)

# Crear un modelo de regresión lineal
model = LinearRegression()

# Entrenar el modelo
model.fit(months.reshape(-1, 1), precio_accion)

# Generar datos para la línea de regresión
future_months = np.arange(1, 205)  # Hasta el año 2027 (12*17 = 204 meses)
predicted_prices = model.predict(future_months.reshape(-1, 1))

# Graficar los datos de ejemplo y la línea de regresión
plt.scatter(months, precio_accion, color='blue', label='Datos de ejemplo')
plt.plot(future_months, np.maximum(2, predicted_prices), color='red', label='Línea de regresión')
plt.scatter([future_months[-1]], [np.maximum(2, predicted_prices[-1])], color='green', label='Predicción 2027')
plt.xlabel('Mes')
plt.ylabel('Precio de la acción')
plt.title('Comportamiento del precio de una acción desde 2010 hasta 2023')
plt.legend()
plt.grid(True)
plt.show()

# Hacer una predicción para el año 2027
meses_2027 = (2027 - 2010) * 12  # Convertir años a meses
prediccion_2027 = model.predict([[meses_2027]])
print(f"Según el modelo de regresión lineal, el precio de la acción en el año 2027 sería de aproximadamente: ${np.maximum(2, prediccion_2027[0]):.2f}")






```
